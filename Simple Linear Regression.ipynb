## Simple Linear Regression Using Numpy and Scikit-Learn


### Load Packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

Create random temperature values over some mins period.

```python
mins = np.arange(30) 
delta = np.random.uniform(32, 45, size=(mins.shape[0], ))
temps = .2 * mins + delta
```

```python
# Plot
fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(mins, temps, label='measured')
ax.set_ylabel('Temp')
ax.set_xlabel('Mins')
```

## Using Numpy's Polyfit for Linear Fitting

```python
slope, intercept = np.polyfit(mins, temps, deg=1) # set deg 1 for linear.
# calculate fitted line
fit = intercept + slope*mins

# Plot fitted line
fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(mins, temps, label='measured')
ax.plot(mins, fit, 'r-', label='fit')
ax.set_ylabel('Temp')
ax.set_xlabel('Mins')
plt.legend()
plt.show()
```

## Fitting with Scikit-Learn

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
regression = lr.fit(X=mins.reshape(-1, 1), y=temps.reshape(-1, 1))
regression.intercept_, regression.coef_
```

```python
# Calculate Fitted line
fit2 = regression.intercept_ + regression.coef_*mins
```

```python
# Plot fitted line
fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(mins, temps, label='measured')
ax.plot(mins.ravel(), fit2.ravel(), 'r-', label='fit')
plt.legend()
plt.show()
```
